```python
# Perform the t-test between mainstream vs premium and budget midage & young singles and couples

import pandas as pd
from scipy.stats import ttest_ind
```

```python
segments = ['YOUNG SINGLES/COUPLES', 'MIDAGE SINGLES/COUPLES']

filtered_df = df_trans_data[df_trans_data['LIFESTAGE'].isin(segments)]

# Separate the data into mainstream and premium segments

mainstream = filtered_df[filtered_df['PREMIUM_CUSTOMER'] == 'Mainstream']['PROD_QTY']
premium = filtered_df[filtered_df['PREMIUM_CUSTOMER'] == 'Premium']['PROD_QTY']
```

```python

t_stat, p_value = ttest_ind(mainstream, premium, equal_var=False) 

print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The difference in average units sold between mainstream and premium segments is statistically significant.")
else:
    print("The difference in average units sold between mainstream and premium segments is not statistically significant.")
```
